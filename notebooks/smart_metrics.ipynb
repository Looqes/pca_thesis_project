```python
def create_mean_metrics_table(metrics_df):
    filtered_metrics_df = metrics_df[["filename", 
                            "structure", 
                            "volumetric_dice", 
                            "hausdorff_distance", 
                            # "mean_distance_1_to_2", 
                            # "mean_distance_2_to_1", 
                            "surface_dice_1.0", 
                            "surface_dice_2.0", 
                            "surface_dice_3.0", 
                            # "surface_dice_5.0"
                            ]]
    
    targets = filtered_metrics_df["structure"].unique()
    result = pd.DataFrame()
    
    for target in targets:
        result[target] = filtered_metrics_df[filtered_metrics_df["structure"] == target] \
                         .drop(["filename", "structure"], axis=1) \
                         .mean(axis = 0)

    return result
```

```python
def capitalize_first_two_words(sentence):
    output = ""
    split_sentence = sentence.split(" ")

    for i, word in enumerate(split_sentence):
        if i < 2:
            output += word.capitalize()
        else:
            output += word
        
        output += " " if i < len(split_sentence) else None

    return output

def format_mean_metrics_df(mean_metrics_df):
    mean_metrics_df.index = [capitalize_first_two_words(item.replace("_", " ").replace(".0", " mm")) for item in mean_metrics_df.index]
    for column in mean_metrics_df.columns:
        mean_metrics_df[column] = mean_metrics_df[column].apply(lambda x: round(x, 3))


def combine_model_dataframes(name_df_pairs):
    for name, df in name_df_pairs:
        df.columns = pd.MultiIndex.from_tuples([(name, column) for column in df.columns])

    return pd.concat([df for _, df in name_df_pairs], axis = 1)

```

```python
def combine_model_dataframes(name_df_pairs):
    region_names = ["GG3", "GG4", "Cribriform"]
    model_names = [name for name, _ in name_df_pairs]

    region_frames = {}

    for region_name in region_names:
        region_frames[region_name] = pd.DataFrame(
            [df[1][region_name] for df in name_df_pairs]
        ).T

    # for region in region_frames:
    #     display(region_frames[region])


    for region_name in region_names:
        region_frame = region_frames[region_name]
        multi_index = pd.MultiIndex.from_tuples([(region_name, model_name) for model_name in model_names])
        region_frame.columns = multi_index
        # display(region_frame)

    return pd.concat([df for df in [region_frames[region] for region in region_frames]], axis = 1)
```

Significance test for difference between the all modalities model and t2_adc model

```python
from scipy.stats import ks_2samp
```

```python

# Dice-only data
all_modalities_set = metrics_dfs[0][["filename", "structure", "volumetric_dice"]]
t2_adc_set = metrics_dfs[1][["filename", "structure", "volumetric_dice"]]
```

```python
gg3_data_am = all_modalities_set[all_modalities_set["structure"] == "GG3"]
gg3_data_t2adc = t2_adc_set[t2_adc_set["structure"] == "GG3"]

display(gg3_data_am.head(5))
display(gg3_data_t2adc.head(5))
```

```python
# sns.histplot(all_modalities_set["volumetric_dice"])
sns.histplot(gg3_data_am["volumetric_dice"])
sns.histplot(gg3_data_t2adc["volumetric_dice"])
```

```python

# Generate two random samples (replace with your actual data)
# x = np.random.normal(0, 1, 1000)
# z = np.random.normal(1.1, 0.9, 1000)




for pattern in ["GG3", "GG4", "Cribriform"]: 
    print(f"{pattern}")
    x = all_modalities_set[all_modalities_set["structure"] == pattern]["volumetric_dice"]
    y = t2_adc_set[t2_adc_set["structure"] == pattern]["volumetric_dice"]


    # Perform the two-sample Kolmogorov-Smirnov test
    ks_statistic, p_value = ks_2samp(x, y)

    # print(ks_statistic, p_value)
    print("P-value = ", p_value)

    # Interpret the results
    if p_value < 0.05:
        print("The two distributions are significantly different.")
    else:
        print("Cannot reject the null hypothesis that the two distributions are identical.")
    
    print("---\n")
```
