```python
def create_mean_metrics_table(metrics_df):
    filtered_metrics_df = metrics_df[["filename", 
                            "structure", 
                            "volumetric_dice", 
                            "hausdorff_distance", 
                            "mean_distance_1_to_2", 
                            "mean_distance_2_to_1", 
                            "surface_dice_1.0", 
                            "surface_dice_2.0", 
                            "surface_dice_3.0", 
                            "surface_dice_5.0"]]

    GG3_metrics = filtered_metrics_df[filtered_metrics_df["structure"] == "GG3"]
    GG4_metrics = filtered_metrics_df[filtered_metrics_df["structure"] == "GG4"]
    Cribriform_metrics = filtered_metrics_df[filtered_metrics_df["structure"] == "Cribriform"]
    GG3_mean_metrics = GG3_metrics.drop(["filename", "structure"], axis=1).mean(axis = 0)
    GG4_mean_metrics = GG4_metrics.drop(["filename", "structure"], axis=1).mean(axis = 0)
    Cribriform_mean_metrics = Cribriform_metrics.drop(["filename", "structure"], axis=1).mean(axis = 0)

    return pd.DataFrame({"GG3": GG3_mean_metrics, "GG4": GG4_mean_metrics, "Cribriform": Cribriform_mean_metrics})

all_modalities_mean_metrics = create_mean_metrics_table(metrics_df_all_modalities)

```

```python
def capitalize_first_two_words(sentence):
    output = ""
    split_sentence = sentence.split(" ")

    for i, word in enumerate(split_sentence):
        if i < 2:
            output += word.capitalize()
        else:
            output += word
        
        output += " " if i < len(split_sentence) else None

    return output

def format_mean_metrics_df(mean_metrics_df):
    mean_metrics_df.index = [capitalize_first_two_words(item.replace("_", " ").replace(".0", " mm")) for item in mean_metrics_df.index]
    mean_metrics_df["GG3"] = mean_metrics_df["GG3"].apply(lambda x: round(x, 3))
    mean_metrics_df["GG4"] = mean_metrics_df["GG4"].apply(lambda x: round(x, 3))
    mean_metrics_df["Cribriform"] = mean_metrics_df["Cribriform"].apply(lambda x: round(x, 3))

format_mean_metrics_df(all_modalities_mean_metrics)

print(all_modalities_mean_metrics)
```

```python
t2_adc_mean_metrics = create_mean_metrics_table(metrics_df_t2_adc)
format_mean_metrics_df(t2_adc_mean_metrics)

t2_perfusion_mean_metrics = create_mean_metrics_table(metrics_df_t2_perfusion)
format_mean_metrics_df(t2_perfusion_mean_metrics)
```
